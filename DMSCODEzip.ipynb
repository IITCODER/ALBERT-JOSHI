# **Basic logical operations**

```python
import itertools

# ----- Basic logical operations -----
def NOT(p): return not p
def AND(p, q): return p and q
def OR(p, q): return p or q
def IMPLIES(p, q): return (not p) or q
def IFF(p, q): return p == q

# ----- Truth table generator -----
def truth_table(expr, variables):
    print("\nTruth Table for:", expr.__name__)
    header = " | ".join(variables) + " | Result"
    print(header)
    print("-" * len(header))
    for values in itertools.product([False, True], repeat=len(variables)):
        env = dict(zip(variables, values))
        result = expr(**env)
        row = " | ".join(str(int(env[v])) for v in variables) + " | " + str(int(result))
        print(row)

# ----- Classification: tautology / contradiction / contingency -----
def classify(expr, variables):
    results = []
    for values in itertools.product([False, True], repeat=len(variables)):
        env = dict(zip(variables, values))
        results.append(expr(**env))
    if all(results):
        return "Tautology"
    elif not any(results):
        return "Contradiction"
    else:
        return "Contingency"

# ----- Equivalence check -----
def equivalent(expr1, expr2, variables):
    for values in itertools.product([False, True], repeat=len(variables)):
        env = dict(zip(variables, values))
        if expr1(**env) != expr2(**env):
            return False
    return True

# ----- Predicate logic demo -----
def predicate_logic(domain):
    P = lambda x: x % 2 == 0  # "x is even"
    Q = lambda x: x > 1       # "x > 1"

    print("\nPredicate logic on domain:", domain)
    print("x (P(x)  Q(x)) =", all(IMPLIES(P(x), Q(x)) for x in domain))
    print("x (P(x)  Q(x)) =", any(AND(P(x), Q(x)) for x in domain))

# ================= MAIN PROGRAM =================

# Example statements
expr1 = lambda p, q: IMPLIES(p, q)       # p  q
expr2 = lambda p, q: OR(NOT(p), q)       # p  q

# Show truth table
truth_table(expr1, ["p", "q"])
print("Classification:", classify(expr1, ["p", "q"]))

# Check equivalence
print("\nEquivalence of (p  q) and (p  q):", equivalent(expr1, expr2, ["p", "q"]))

# Run predicate logic example
predicate_logic([1, 2, 3, 4, 5])

# Interactive input
print("\n--- Interactive Demo ---")
print("Enter values of p and q (0=False, 1=True):")
p = bool(int(input("p = ")))
q = bool(int(input("q = ")))
print(f"p AND q = {AND(p,q)}")
print(f"p OR q  = {OR(p,q)}")
print(f"p  q   = {IMPLIES(p,q)}")
print(f"p  q   = {IFF(p,q)}")
```

```python
import itertools

# ----- Basic logical operations -----
def NOT(p): return not p
def AND(p, q): return p and q
def OR(p, q): return p or q
def IMPLIES(p, q): return (not p) or q
def IFF(p, q): return p == q

# ----- Truth table generator -----
def truth_table(expr, variables):
    print("\nTruth Table for:", expr.__name__)
    header = " | ".join(variables) + " | Result"
    print(header)
    print("-" * len(header))
    for values in itertools.product([False, True], repeat=len(variables)):
        env = dict(zip(variables, values))
        result = expr(**env)
        row = " | ".join(str(int(env[v])) for v in variables) + " | " + str(int(result))
        print(row)

# ----- Classification: tautology / contradiction / contingency -----
def classify(expr, variables):
    results = []
    for values in itertools.product([False, True], repeat=len(variables)):
        env = dict(zip(variables, values))
        results.append(expr(**env))
    if all(results):
        return "Tautology"
    elif not any(results):
        return "Contradiction"
    else:
        return "Contingency"

# ----- Equivalence check -----
def equivalent(expr1, expr2, variables):
    for values in itertools.product([False, True], repeat=len(variables)):
        env = dict(zip(variables, values))
        if expr1(**env) != expr2(**env):
            return False
    return True

# ----- Predicate logic demo -----
def predicate_logic(domain):
    P = lambda x: x % 2 == 0  # "x is even"
    Q = lambda x: x > 1       # "x > 1"

    print("\nPredicate logic on domain:", domain)
    print("x (P(x)  Q(x)) =", all(IMPLIES(P(x), Q(x)) for x in domain))
    print("x (P(x)  Q(x)) =", any(AND(P(x), Q(x)) for x in domain))

# ================= MAIN PROGRAM =================

# Example statements
expr1 = lambda p, q: IMPLIES(p, q)       # p  q
expr2 = lambda p, q: OR(NOT(p), q)       # p  q
expr3 = lambda p, q, r: AND(p, OR(q, r)) # p  (q  r)

# Show truth tables
truth_table(expr1, ["p", "q"])
print("Classification:", classify(expr1, ["p", "q"]))

truth_table(expr3, ["p", "q", "r"])
print("Classification:", classify(expr3, ["p", "q", "r"]))

# Check equivalence
print("\nEquivalence of (p  q) and (p  q):", equivalent(expr1, expr2, ["p", "q"]))

# Run predicate logic example
predicate_logic([1, 2, 3, 4, 5])

# Interactive input for multiple variables
print("\n--- Interactive Demo ---")
vars_list = ["p", "q", "r"]
values = {}
for v in vars_list:
    values[v] = bool(int(input(f"{v} (0=False, 1=True): ")))

print("\nResults with your inputs:")
print(f"p AND q = {AND(values['p'], values['q'])}")
print(f"p OR q  = {OR(values['p'], values['q'])}")
print(f"(p  q) = {IMPLIES(values['p'], values['q'])}")
print(f"(p  q) = {IFF(values['p'], values['q'])}")
print(f"p  (q  r) = {AND(values['p'], OR(values['q'], values['r']))}")
```

# **WELL FORMED FORMULA**

```python
# Simple WFF (Well-Formed Formula) checker

def is_wff(formula):
    allowed_vars = set("pqrxyz")   # allowed variables
    allowed_ops = set("()")    # allowed operators
    stack = []

    for char in formula:
        if char.isalpha() and char not in allowed_vars:
            return False, f"Invalid variable: {char}"
        elif char not in allowed_vars and char not in allowed_ops:
            return False, f"Invalid symbol: {char}"
        if char == "(":
            stack.append(char)
        elif char == ")":
            if not stack:
                return False, "Unmatched closing parenthesis"
            stack.pop()

    if stack:
        return False, "Unmatched opening parenthesis"
    return True, "Well-formed formula"


# -----------------------------
# Examples
# -----------------------------
formulas = [
    "(p  q)",       # valid
    "p  (q  r)",  # valid
    "p  (q  )",    # invalid
    "(p  q"         # invalid
]

for f in formulas:
    ok, msg = is_wff(f)
    print(f"{f} -> {msg}")
```

# **Tautoloy**

```python
import itertools
import re

# Evaluate formula safely with truth values
def evaluate(expr, values):
    # Perform replacements carefully to maintain valid syntax
    expr = expr.replace("<->", "==")  # IFF
    expr = expr.replace("->", " or (not ") # IMPLIES (correctly forms 'a or (not b)')
    expr = expr.replace("~", " not ") # NOT
    expr = expr.replace("&", " and ") # AND
    expr = expr.replace("|", " or ")  # OR


    # Replace variables with truth values
    for var, val in values.items():
        # Use regex to ensure whole word replacement, avoiding issues with variable names like 'p' being part of 'apply'
        expr = re.sub(rf"\b{var}\b", str(val), expr)
    try:
        # Evaluate the translated expression. Using eval is powerful but requires care.
        # We are evaluating a string that should now be valid Python boolean logic.
        return eval(expr)
    except SyntaxError as e:
        print(f"Syntax Error during evaluation of expression: {expr}")
        raise e # Re-raise the exception after printing the problematic expression
    except Exception as e:
        print(f"Error evaluating expression: {expr}")
        raise e


# Detect variables automatically
def extract_variables(formula):
    return sorted(list(set(re.findall(r"[a-z]", formula)))) # Assumes variables are single lowercase letters


# Print truth table + check tautology
def truth_table(formula):
    variables = extract_variables(formula)
    if not variables and formula.strip(): # Handle formulas with no variables, like "(True | False)"
        print("\nEvaluating constant formula:", formula)
        try:
            result = eval(formula.replace("True", "True").replace("False", "False").replace("~", " not ").replace("&", " and ").replace("|", " or "))
            print(f"Result: {result}")
            print("-" * 20)
            if result:
                 print("The formula is a TAUTOLOGY")
            else:
                 print(" The formula is a CONTRADICTION")
        except Exception as e:
            print(f"Error evaluating constant formula: {e}")
        return


    if not variables:
        print("No variables found in the formula.")
        return


    print("\nTruth Table for:", formula)
    # Adjust header width based on number of variables
    header_width = len(" | ".join(variables)) + len(" | Formula")
    print("-" * (header_width + 2 * len(variables) )) # Added padding based on number of variables
    print(" | ".join(v.ljust(5) for v in variables) + " | Formula") # Align variable names
    print("-" * (header_width + 2 * len(variables) ))

    results = []
    for assignment in itertools.product([True, False], repeat=len(variables)):
        values = dict(zip(variables, assignment))
        try:
            result = evaluate(formula, values)
            results.append(result)
            row = " | ".join(str(values[v]).ljust(5) for v in variables)
            print(f"{row} | {str(result).ljust(7)}") # Align result column
        except Exception as e:
            print(f"Error processing row for values {values}: {e}")
            # Depending on desired behavior, you might want to break or continue
            return # Stop table generation on error


    print("-" * (header_width + 2 * len(variables) ))
    if all(results):
        print("The formula is a TAUTOLOGY")
    elif not any(results):
        print(" The formula is a CONTRADICTION")
    else:
        print(" The formula is CONTINGENT (sometimes true, sometimes false)")

# -----------------------------
# Example usage
# -----------------------------
formulas = [
    "(p | ~p)",                     # tautology
    "(p & q) -> p",                 # tautology
    "(p & q)",                      # not tautology
    "(p -> q) <-> (~q -> ~p)",       # tautology
    "p -> (q -> p)",              # Tautology
    "((p -> q) & p) -> q",        # Tautology (Modus Ponens)
    "(p | q) & (~p | r) -> (q | r)" # Tautology (Resolution)
]

for f in formulas:
    truth_table(f)
```

# **Equivalence Formula**

```python
import itertools
import re

# Evaluate a formula safely with truth values
def evaluate(expr, values):
    expr = expr.replace("->", ") or (not ") if "->" in expr else expr
    expr = expr.replace("<->", "==")
    expr = expr.replace("~", " not ")
    expr = expr.replace("&", " and ")
    expr = expr.replace("|", " or ")

    # Replace variables with truth values
    for var, val in values.items():
        expr = re.sub(rf"\b{var}\b", str(val), expr)
    return eval(expr)

# Detect variables automatically
def extract_variables(formula):
    return sorted(set(re.findall(r"[a-z]", formula)))

# Check equivalence of two formulas
def are_equivalent(f1, f2):
    variables = sorted(set(extract_variables(f1)) | set(extract_variables(f2)))

    print(f"\nTruth Table for Equivalence Check:\nFormula 1: {f1}\nFormula 2: {f2}")
    print("-" * (len(variables) * 8 + 20))
    print(" | ".join(variables) + " | F1 | F2 ")
    print("-" * (len(variables) * 8 + 20))

    equivalent = True
    for assignment in itertools.product([True, False], repeat=len(variables)):
        values = dict(zip(variables, assignment))
        val1 = evaluate(f1, values)
        val2 = evaluate(f2, values)
        row = " | ".join(str(values[v]).ljust(5) for v in variables)
        print(f"{row} | {val1} | {val2}")
        if val1 != val2:
            equivalent = False

    print("-" * (len(variables) * 8 + 20))
    if equivalent:
        print("The two formulas are EQUIVALENT")
    else:
        print("The two formulas are NOT equivalent")

# -----------------------------
# Example usage
# -----------------------------
are_equivalent("(p -> q)", "(~q -> ~p)")       # Contrapositive
are_equivalent("(p | q)", "(q | p)")           # Commutativity
are_equivalent("(p & q)", "(p | q)")           # Not equivalent
```

# **Predicates and Quantifiers**

```python
# Predicates and Quantifiers in Python

# Universal Quantifier ()
def universal(domain, predicate):
    return all(predicate(x) for x in domain)

# Existential Quantifier ()
def existential(domain, predicate):
    return any(predicate(x) for x in domain)

# -----------------------------
# Example Predicates
# -----------------------------
# Predicate: "x is even"
is_even = lambda x: x % 2 == 0

# Predicate: "x is greater than 5"
greater_than_5 = lambda x: x > 5

# Domain
domain = range(1, 11)  # {1,2,...,10}

# -----------------------------
# Examples
# -----------------------------
print("Domain:", list(domain))

# x P(x) : all elements satisfy P
print("x (x is even) ?", universal(domain, is_even))
print("x (x > 5) ?", universal(domain, greater_than_5))

# x P(x) : at least one element satisfies P
print("x (x is even) ?", existential(domain, is_even))
print("x (x > 5) ?", existential(domain, greater_than_5))
```

# **Hasse diagram**
image.png

```python
import networkx as nx
import matplotlib.pyplot as plt

def is_partial_order(A, R):
    A = set(A)
    R = set(R)

    # Reflexive
    reflexive = all((a, a) in R for a in A)

    # Antisymmetric
    antisymmetric = all((a == b) or ((b, a) not in R) for (a, b) in R)

    # Transitive
    transitive = all(((a, c) in R) for (a, b) in R for (x, c) in R if x == b)

    return reflexive and antisymmetric and transitive


def hasse_diagram(A, R):
    if not is_partial_order(A, R):
        print(" The relation is NOT a partial order. Hasse diagram not applicable.")
        return

    # Step 1: Start with the relation
    H = set(R)

    # Step 2: Remove reflexive edges
    H = {(a, b) for (a, b) in H if a != b}

    # Step 3: Remove transitive edges
    for a in A:
        for b in A:
            for c in A:
                if (a, b) in H and (b, c) in H and (a, c) in H:
                    H.remove((a, c))

    # Step 4: Draw Hasse Diagram
    G = nx.DiGraph()
    G.add_nodes_from(A)
    G.add_edges_from(H)

    pos = nx.spring_layout(G)  # layout for visualization
    plt.figure(figsize=(5,5))
    nx.draw(G, pos, with_labels=True, node_size=800, node_color="lightgreen",
            font_size=12, font_weight="bold", arrows=True, arrowsize=20)
    plt.title("Hasse Diagram of Partial Order")
    plt.show()

    print(" Hasse diagram drawn for relation:", H)


# -----------------------------
# Example Usage
# -----------------------------
A = {1, 2, 3}
R = {(1,1), (2,2), (3,3), (1,2), (2,3), (1,3)}  # Partial order

hasse_diagram(A, R)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def is_partial_order(A, R):
    A, R = set(A), set(R)
    reflexive = all((a, a) in R for a in A)
    antisymmetric = all((a == b) or ((b, a) not in R) for (a, b) in R)
    transitive = all(((a, c) in R) for (a, b) in R for (x, c) in R if x == b)
    return reflexive and antisymmetric and transitive

def hasse_edges(A, R):
    """Compute cover relation (remove reflexive + transitive edges)."""
    H = {(a, b) for (a, b) in R if a != b}
    for a in A:
        for b in A:
            for c in A:
                if (a, b) in H and (b, c) in H and (a, c) in H:
                    H.remove((a, c))
    return H

def leq(a, b, R):
    return (a, b) in R

def meet(a, b, A, R):
    """Greatest lower bound of a and b."""
    lowers = [x for x in A if leq(x,a,R) and leq(x,b,R)]
    glb = [x for x in lowers if all(not leq(x,y,R) or x==y for y in lowers)]
    return glb[0] if glb else None

def join(a, b, A, R):
    """Least upper bound of a and b."""
    uppers = [x for x in A if leq(a,x,R) and leq(b,x,R)]
    lub = [x for x in uppers if all(not leq(y,x,R) or x==y for y in uppers)]
    return lub[0] if lub else None

def is_lattice(A, R):
    if not is_partial_order(A, R):
        return False
    for a in A:
        for b in A:
            if meet(a, b, A, R) is None or join(a, b, A, R) is None:
                return False
    return True

def draw_hasse(A, R, title="Hasse Diagram"):
    H = hasse_edges(A, R)
    G = nx.DiGraph()
    G.add_nodes_from(A)
    G.add_edges_from(H)

    pos = nx.spring_layout(G)
    plt.figure(figsize=(6,6))
    nx.draw(G, pos, with_labels=True, node_size=800, node_color="lightblue",
            font_size=12, font_weight="bold", arrows=True, arrowsize=20)
    plt.title(title)
    plt.show()

# -----------------------------
# Example Usage
# -----------------------------
A = {1, 2, 3, 6}
R = {
    (1,1),(2,2),(3,3),(6,6),
    (1,2),(1,3),(1,6),(2,6),(3,6)
}  # divisibility relation on {1,2,3,6}

print("Set A:", A)
print("Relation R:", R)

if is_lattice(A, R):
    print(" The relation is a LATTICE")
    draw_hasse(A, R, "Hasse Diagram of Lattice")
else:
    print(" The relation is NOT a lattice")
```

```python
#DMS CODING CLASS IN PYTHON
from itertools import product
# Example: Z2 x Z2 x Z2 GROUP UNDER ADDITION ?
Z2 = [0, 1]
elements_Z2xZ2xZ2 = list(product(Z2, Z2, Z2))
print("Z2 x Z2 x Z2:")
print(elements_Z2xZ2xZ2)

# Example: Z2 x Z4 GROUP UNDER ADDITION ?
Z4 = [0, 1, 2, 3]
elements_Z2xZ4 = list(product(Z2, Z4))
print("\nZ2 x Z4:")
print(elements_Z2xZ4)

# Example: Z8 GROUP UNDER ADDITION ?
Z8 = list(range(8))
print("\nZ8:")
print(Z8)
```

```python
#DMS CODING CLASS IN PYTHON
from itertools import product
# Define Z2 x Z4
Z2 = [0, 1]
Z4 = [0, 1, 2, 3]
G = list(product(Z2, Z4))

# Addition modulo (2,4)
def add(a, b):
    return ((a[0] + b[0]) % 2, (a[1] + b[1]) % 4)

# 1. Closure
closure = all(add(a, b) in G for a in G for b in G)

# 2. Associativity
associativity = all(add(add(a, b), c) == add(a, add(b, c)) for a in G for b in G for c in G)

# 3. Identity
identity_candidates = [e for e in G if all(add(a, e) == a and add(e, a) == a for a in G)]
identity = identity_candidates[0] if identity_candidates else None

# 4. Inverses
inverses = {a: [b for b in G if add(a, b) == identity] for a in G}
all_inverses_exist = all(len(v) > 0 for v in inverses.values())

# Results
print("Z2 x Z4 as a group under addition:")
print("Closure holds:", closure)
print("Associativity holds:", associativity)
print("Identity element:", identity)
print("All elements have inverses:", all_inverses_exist)
print("\nInverses of elements:")
for a, invs in inverses.items():
    print(f"{a} -> {invs}")
```

1. **`for a in G for b in G`**
    This creates all possible pairs $(a, b)$ from the group $G$.
   (In our case, $G = \mathbb{Z}_2 \times \mathbb{Z}_4$, which has 8 elements, so it generates $8 \times 8 = 64$ pairs.)

2. **`add(a, b) in G`**
    For each pair, it checks whether the sum (using our `add` function with mod 2 and mod 4) is still inside the set $G$.

3. **`all(...)`**
    The built-in Python function `all()` returns **True** if **every condition** inside is True; otherwise it returns **False**.
