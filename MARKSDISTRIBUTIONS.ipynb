```python
#MARKS DISTRIBUTIONS MACHINE LEARNING CODE
# Author: Yogash Chhandra Joshi
import pandas as pd
import matplotlib.pyplot as plt

# ========================
# Load dataset
# ========================
file_path = "MARKS.xlsx"

# Load first sheet
df = pd.read_excel(file_path, sheet_name=0)

# Drop empty columns
df = df.dropna(axis=1, how='all')

print("Dataset Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nFirst 5 rows:\n", df.head())

# ========================
# Define attribute groups
# Author: Yogash Chhandra Joshi
# ========================
group1 = ["A","B","C","D","E","F","G","H","I","J"]   # first group 1 Marks
group2 = ["B1","B2","B3","B4","B5","B6"]                # second group 5 Marks

# Make sure they exist in dataset
group1 = [c for c in group1 if c in df.columns]
group2 = [c for c in group2 if c in df.columns]

print("\nAvailable Group1 features:", group1)
print("Available Group2 features:", group2)

# ========================
# Helper function
# Author: Yogash Chhandra Joshi
# ========================
def analyze_features(data, features, title_group):
    """Generate descriptive stats, histograms, and bar charts"""
    for col in features:
        # Convert to numeric
        data[col] = pd.to_numeric(data[col], errors="coerce")

        print(f"\n--- Analysis for {col} ---")
        print("Mean:", data[col].mean())
        print("Median:", data[col].median())
        print("Std Dev:", data[col].std())
        print("Min:", data[col].min(), "Max:", data[col].max())

        # Histogram
        plt.figure(figsize=(6,4))
        plt.hist(data[col].dropna(), bins=10, edgecolor="black")
        plt.title(f"{title_group}: Distribution of {col}")
        plt.xlabel("Marks")
        plt.ylabel("Number of Students")
        plt.grid(axis="y", linestyle="--", alpha=0.7)
        plt.show()

        # Bar (average)
        plt.figure(figsize=(4,4))
        plt.bar([col], [data[col].mean()], color="deeppink", edgecolor="black")
        plt.title(f"{title_group}: Average of {col}")
        plt.ylabel("Average Marks")
        plt.show()

# ========================
# Run analysis
# Author: Yogash Chhandra Joshi
# ========================
analyze_features(df, group1, "Group1 (A–J)")
analyze_features(df, group2, "Group2 (B1–B6)")

# ========================
# Final FULL MARKS analysis
# Author: Yogash Chhandra Joshi
# ========================
if "FINAL MARKS" in df.columns:
    df["FINAL MARKS"] = pd.to_numeric(df["FINAL MARKS"], errors="coerce")

    print("\n--- FINAL ANALYSIS: FINAL MARKS ---")
    print(df["FINAL MARKS"].describe())

    # Scatter (student vs full marks)
    student_col = df.columns[0]  # assuming first column = student ID/roll
    plt.figure(figsize=(10,5))
    plt.scatter(df[student_col], df["FINAL MARKS"], color="dodgerblue", alpha=0.7)
    plt.title("FINAL Marks per Student")
    plt.xlabel("Student")
    plt.ylabel("FINAL Marks")
    plt.xticks(rotation=90)
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

    # Pie chart (pass vs fail, assuming 40% passing)
    max_marks = df["FINAL MARKS"].max()
    pass_mark = 0.4 * max_marks
    status = df["FINAL MARKS"].apply(lambda x: "Pass" if x >= pass_mark else "Fail")

    plt.figure(figsize=(6,6))
    status.value_counts().plot.pie(autopct="%1.1f%%", colors=["lawngreen","salmon"], startangle=90)
    plt.title("Pass vs Fail Distribution (FINAL MARKS)")
    plt.ylabel("")
    plt.show()
```

# **Hypothesis Testing**

```python
from scipy import stats
# ========================
# Hypothesis Testing on FINAL MARKS
# Author: Yogash Chhandra Joshi
# ========================
if "FINAL MARKS" in df.columns:
    final_marks = pd.to_numeric(df["FINAL MARKS"], errors="coerce").dropna()

    # Hypothesis: mean = 15 (out of 30)
    mu = 15

    # One-sample t-test
    t_stat, p_val = stats.ttest_1samp(final_marks, mu)

    print("\n=== Hypothesis Testing: FINAL MARKS ===")
    print(f"Null Hypothesis H0: mean(FINAL MARKS) = {mu}")
    print(f"Alternative Hypothesis H1: mean(FINAL MARKS) ≠ {mu}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_val:.4f}")

    alpha = 0.05
    if p_val < alpha:
        print("Reject H0 → The mean FINAL MARKS is significantly different from", mu)
    else:
        print("Fail to Reject H0 → The mean FINAL MARKS is not significantly different from", mu)
```

# **T-statistic**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Ensure FINAL MARKS is numeric
final_marks = pd.to_numeric(df["FINAL MARKS"], errors="coerce").dropna()

# Parameters
mu0 = 15  # Null hypothesis mean
alpha = 0.05
n = len(final_marks)
sample_mean = final_marks.mean()
sample_std = final_marks.std(ddof=1)

# Standard error
se = sample_std / np.sqrt(n)

# Test statistic
t_stat = (sample_mean - mu0) / se
dfree = n - 1

print("Sample Mean:", sample_mean)
print("T-statistic:", t_stat)

# ======================
# Left-tailed Test Graph
# Author: Yogash Chhandra Joshi
# ======================
x = np.linspace(-4, 4, 500)
y = stats.t.pdf(x, dfree)

plt.figure(figsize=(8,5))
plt.plot(x, y, 'b-', label='t-distribution (df={})'.format(dfree))

# Critical value
t_crit_left = stats.t.ppf(alpha, dfree)

# Shade rejection region (left)
plt.fill_between(x, 0, y, where=(x < t_crit_left), color="red", alpha=0.5, label="Rejection Region")

# Mark observed t
plt.axvline(t_stat, color="green", linestyle="--", linewidth=2, label=f"T-stat = {t_stat:.2f}")

plt.title("Left-Tailed Test (H1: μ < 15)")
plt.xlabel("t-value")
plt.ylabel("Density")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# ======================
# Right-tailed Test Graph
# Author: Yogash Chhandra Joshi
# ======================
plt.figure(figsize=(8,5))
plt.plot(x, y, 'b-', label='t-distribution (df={})'.format(dfree))

# Critical value
t_crit_right = stats.t.ppf(1-alpha, dfree)

# Shade rejection region (right)
plt.fill_between(x, 0, y, where=(x > t_crit_right), color="red", alpha=0.5, label="Rejection Region")

# Mark observed t
plt.axvline(t_stat, color="green", linestyle="--", linewidth=2, label=f"T-stat = {t_stat:.2f}")

plt.title("Right-Tailed Test (H1: μ > 15)")
plt.xlabel("t-value")
plt.ylabel("Density")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# **Density Curve of FINAL MARKS with Mean Comparison**

###  Explanation

1. A **density curve** is not the same as a frequency histogram.

   * In a histogram, the **y-axis shows counts** (how many students got marks).
   * In a density plot, the **y-axis shows probability density**, and the **area under the curve = 1** (100%).

2. Since your marks range from **0 to 30**, the curve must "spread out" across 30 units.
   To make the total area = 1, the **height of the curve is relatively small**.
   That’s why the peak density is around **0.05** (≈ 5%).

   **Mathematically:**

   $$
   \text{Height of peak} \approx \frac{1}{\text{spread of marks}}
   $$

   Since spread ≈ 30, the peak ≈ 1/30 ≈ 0.033 → close to what you see.


###  How to interpret

 The y-axis is not “number of students” but “probability density.”

 If the density is **0.05 at x = 20**, it means the likelihood of a student scoring \~20 is higher than other regions.

 The total shaded area under the curve = 1 (all students).

# **DEEP LEARNING FOR MARKS DISTRIBUTIONS**

```python
#OUR MARKS DISTRIBUTION IN FINAL MARKS ACCORDING TO DEEP LEARNING
# Author: Yogash Chhandra Joshi
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


# Plot clusters + deep learning regression
plt.figure(figsize=(12,7))

colors = {
    "Cluster 1 (0-10)": "red",
    "Cluster 2 (11-20)": "blue",
    "Cluster 3 (21-30)": "green"
}

for cluster, color in colors.items():
    subset = df[df["Cluster"] == cluster]

    if not subset.empty:
        X = subset.index.values.reshape(-1,1)
        y = subset["FINAL MARKS"].values.reshape(-1,1)

        # Normalize data (important for deep learning)
        scaler_X = MinMaxScaler()
        scaler_y = MinMaxScaler()
        X_scaled = scaler_X.fit_transform(X)
        y_scaled = scaler_y.fit_transform(y)

        # Build Neural Network
        model = Sequential([
            Dense(16, activation="relu", input_dim=1),
            Dense(32, activation="relu"),
            Dense(1, activation="linear")   # output layer
        ])

        model.compile(optimizer="adam", loss="mse")
        model.fit(X_scaled, y_scaled, epochs=200, verbose=0)

        # Predict curve
        x_range = np.linspace(X.min(), X.max(), 200).reshape(-1,1)
        x_range_scaled = scaler_X.transform(x_range)
        y_pred_scaled = model.predict(x_range_scaled)
        y_pred = scaler_y.inverse_transform(y_pred_scaled)

        # Scatter plot
        plt.scatter(X, y, label=cluster, color=color, s=80, alpha=0.7, edgecolor="black")

        # Deep learning regression curve
        plt.plot(x_range, y_pred, color=color, linewidth=2, linestyle="-",
                 label=f"{cluster} Deep Learning Fit")

plt.title("Deep Learning Regression (Neural Network) for Clusters", fontsize=14)
plt.xlabel("Student Index", fontsize=12)
plt.ylabel("FINAL MARKS", fontsize=12)
plt.ylim(0, 30)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# **NEURAL NETWORK FOR MARKS DISTRIBUTIONS**

```python
#OUR MARKS DISTRIBUTION IN FINAL MARKS ACCORDING TO Neural Network Regression for FINAL MARKS
# Author: Yogash Chhandra Joshi
# ==============================
#  1. Imports
# Author: Yogash Chhandra Joshi
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# ==============================
# 2. Load Dataset
# Author: Yogash Chhandra Joshi
# ==============================
df = pd.read_excel("MARKS.xlsx")

# Ensure FINAL MARKS is numeric
df["FINAL MARKS"] = pd.to_numeric(df["FINAL MARKS"], errors="coerce")

# Independent variable (X = Student Index)
X = df.index.values.reshape(-1, 1)

# Dependent variable (y = FINAL MARKS)
y = df["FINAL MARKS"].values.reshape(-1, 1)

# ==============================
# 3. Normalize Data
# Author: Yogash Chhandra Joshi
# ==============================
# Neural Networks work better when values are between 0 and 1
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# ==============================
# 4. Build Neural Network Model
# Author: Yogash Chhandra Joshi
# ==============================
model = Sequential([
    Dense(16, activation="relu", input_dim=1),  # hidden layer 1
    Dense(32, activation="relu"),               # hidden layer 2
    Dense(1, activation="linear")               # output layer
])

model.compile(optimizer="adam", loss="mse")

# ==============================
# 5. Train Model
# Author: Yogash Chhandra Joshi
# ==============================
history = model.fit(X_scaled, y_scaled, epochs=200, verbose=0)

# ==============================
# 6. Predict and Inverse Transform
# Author: Yogash Chhandra Joshi
# ==============================
# Predict for the full range of student indices
x_range = np.arange(len(df)).reshape(-1,1)
x_range_scaled = scaler_X.transform(x_range)

y_pred_scaled = model.predict(x_range_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled)

# ==============================
# 7. Visualization
# Author: Yogash Chhandra Joshi
# ==============================
plt.figure(figsize=(10,6))
plt.scatter(X, y, color="blue", alpha=0.6, label="Actual Marks", s=80, edgecolor="black")
plt.plot(x_range, y_pred, color="red", linewidth=2, label="Neural Network Prediction")
plt.title("Neural Network Regression for FINAL MARKS", fontsize=14)
plt.xlabel("Student Index", fontsize=12)
plt.ylabel("FINAL MARKS", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# ==============================
# 8. Evaluate Model
# Author: Yogash Chhandra Joshi
# ==============================
loss = model.evaluate(X_scaled, y_scaled, verbose=0)
print(f"Final Training Loss (MSE): {loss:.4f}")
```

1. **Data Normalization**
   * NNs usually expect input/output between **0 and 1** (or -1 and 1).
   * We use `MinMaxScaler` to scale `Roll No.` and `FINAL MARKS`.
2. **Neural Network Architecture**
   * **Input Layer**: 1 feature (Roll No.)
   * **Hidden Layer 1**: 16 neurons, ReLU activation
   * **Hidden Layer 2**: 32 neurons, ReLU activation
   * **Output Layer**: 1 neuron, Linear activation (since regression problem)
3. **Training**
   * Optimizer = `adam` (adaptive learning)
   * Loss function = `mse` (mean squared error, best for regression)
   * Epochs = 200 (number of passes over data)
4. **Prediction & Inverse Transform**
   * Model predicts **scaled values**, so we convert back to original marks using `inverse_transform`.
5. **Visualization**
   * Blue scatter points = actual marks
   * Red line = NN prediction curve

```python
# Install HuggingFace Transformers
# Author: Yogash Chhandra Joshi
!pip install transformers torch sentencepiece

import pandas as pd
from transformers import pipeline

# ============================
# 1. Load Dataset
# Author: Yogash Chhandra Joshi
# ============================
df = pd.read_excel("MARKS.xlsx")

# ============================
# 2. Convert Marks → Text
# Author: Yogash Chhandra Joshi
# ============================
def describe_student(row):
    return f"Student Roll No {row['Roll No.']} scored {row['FINAL MARKS']} marks out of 30."

df["LLM_Input"] = df.apply(describe_student, axis=1)

# ============================
# 3. Load a Pretrained LLM
# (You can replace with 'google/flan-t5-base', 'tiiuae/falcon-7b-instruct', etc.)
# Author: Yogash Chhandra Joshi
# ============================
generator = pipeline("text2text-generation", model="google/flan-t5-base")

# ============================
# 4. Generate Feedback
# Author: Yogash Chhandra Joshi
# ============================
def generate_feedback(text):
    prompt = f"{text} Provide short feedback about this student’s performance."
    result = generator(prompt, max_length=50, do_sample=True)[0]["generated_text"]
    return result

df["Feedback"] = df["LLM_Input"].apply(generate_feedback)

# ============================
# 5. Generate Overall Summary
# Author: Yogash Chhandra Joshi
# ============================
all_text = " ".join(df["LLM_Input"].tolist())
summary_prompt = f"Here is student performance data: {all_text}. Summarize overall performance."
summary = generator(summary_prompt, max_length=100, do_sample=True)[0]["generated_text"]

# ============================
# 6. Display Results
# Author: Yogash Chhandra Joshi
# ============================
print("\n=== Per Student Feedback ===")
print(df[["Roll No.", "FINAL MARKS", "Feedback"]])

print("\n=== Overall Summary ===")
print(summary)
```

```python
# Install HuggingFace Transformers
# Author: Yogash Chhandra Joshi
!pip install transformers torch sentencepiece

import pandas as pd
from transformers import pipeline

# ============================
# 1. Load Dataset
# Author: Yogash Chhandra Joshi
# ============================
df = pd.read_excel("MARKS.xlsx")

# ============================
# 2. Convert Marks → Text
# Author: Yogash Chhandra Joshi
# ============================
def describe_student(row):
    return f"Student Roll No {row['Roll No.']} scored {row['FINAL MARKS']} marks out of 30."

df["LLM_Input"] = df.apply(describe_student, axis=1)

# ============================
# 3. Load a Pretrained LLM
# Author: Yogash Chhandra Joshi
# ============================
generator = pipeline("text2text-generation", model="google/flan-t5-base")

# ============================
# 4. Generate Feedback per Student
# Author: Yogash Chhandra Joshi
# ============================
def generate_feedback(text):
    prompt = f"{text} Provide short feedback about this student’s performance."
    result = generator(prompt, max_length=50, do_sample=True)[0]["generated_text"]
    return result

df["Feedback"] = df["LLM_Input"].apply(generate_feedback)

# ============================
# 5. Cluster Students
# Author: Yogash Chhandra Joshi
# ============================
def assign_cluster(marks):
    if 0 <= marks <= 10:
        return "Cluster 1 (Weak: 0–10)"
    elif 11 <= marks <= 20:
        return "Cluster 2 (Average: 11–20)"
    elif 21 <= marks <= 30:
        return "Cluster 3 (Strong: 21–30)"
    else:
        return "Out of Range"

df["Cluster"] = df["FINAL MARKS"].apply(assign_cluster)

# ============================
# 6. Generate Cluster Summaries
# Author: Yogash Chhandra Joshi
# ============================
cluster_summaries = {}
for cluster_name, cluster_df in df.groupby("Cluster"):
    cluster_text = " ".join(cluster_df["LLM_Input"].tolist())
    prompt = f"Here is student performance data for {cluster_name}: {cluster_text}. Summarize the performance of this group."
    summary = generator(prompt, max_length=80, do_sample=True)[0]["generated_text"]
    cluster_summaries[cluster_name] = summary

# ============================
# 7. Generate Overall Summary
# Author: Yogash Chhandra Joshi
# ============================
all_text = " ".join(df["LLM_Input"].tolist())
summary_prompt = f"Here is student performance data: {all_text}. Summarize overall performance."
overall_summary = generator(summary_prompt, max_length=100, do_sample=True)[0]["generated_text"]

# ============================
# 8. Display Results
# Author: Yogash Chhandra Joshi
# ============================
print("\n=== Per Student Feedback ===")
print(df[["Roll No.", "FINAL MARKS", "Cluster", "Feedback"]])

print("\n=== Cluster Summaries ===")
for cluster, summ in cluster_summaries.items():
    print(f"{cluster}: {summ}")

print("\n=== Overall Summary ===")
print(overall_summary)
```

# Steps of Hypothesis Testing
Let’s walk through how to do a hypothesis test, one step at a time.


# Step 1: State your hypotheses
The first step is to formulate your research question into two competing hypotheses:

**Null Hypothesis (H0):** This is the default assumption that there is no effect or difference.

**Alternative Hypothesis (Ha):** This is the hypothesis that there is an effect or difference.


For example:

H0: The mean height of men is equal to the mean height of women.

Ha: The mean height of men is not equal to the mean height of women.

# Step 2: Collect and prepare data
Gather data through experiments, surveys, or observational studies. Ensure the data collection method is designed to test the hypothesis and is representative of the population. This step often involves:

Defining the population of interest.
Selecting an appropriate sampling method.
Determining the sample size.
Collecting and organizing the data.


# Step 3: Choose the appropriate statistical test
Select a statistical test based on the type of data and the hypothesis. The choice depends on factors such as:


Data type (continuous, categorical, etc.)
Distribution of the data (normal, non-normal)
Sample size
Number of groups being compared
Common tests include:

t-tests (for comparing means)

chi-square tests (for categorical data)

ANOVA (for comparing means of multiple groups)


# Step 4: Calculate the test statistic and p-value
Use statistical software or formulas to compute the test statistic and corresponding p-value. This step quantifies how much the sample data deviates from the null hypothesis.


The p-value is an important concept in hypothesis testing. It represents the probability of observing results as extreme as the sample data, assuming the null hypothesis is true.


# Step 5: Make a decision
Compare the p-value to the predetermined significance level (α), which is typically set at 0.05. The decision rule is as follows:


If p-value ≤ α: Reject the null hypothesis, suggesting evidence supports the alternative hypothesis.

If p-value > α: Fail to reject the null hypothesis, suggesting insufficient evidence to support the alternative hypothesis.

It's important to note that failing to reject the null hypothesis doesn't prove it's true; it simply means there's not enough evidence to conclude otherwise.

# Step 6: Present your findings
Report the results, including the test statistic, p-value, and conclusion. Discuss whether the findings support the initial hypothesis and their implications. When presenting results, consider:


Providing context for the study.
Clearly stating the hypotheses.
Reporting the test statistic and p-value.
Interpreting the results in plain language.
Discussing the practical significance of the findings.

# **Types of Hypothesis Tests**
Hypothesis tests can be broadly categorized into two main types:

**Parametric tests**
Parametric tests assume that the data follows a specific probability distribution, typically the normal distribution. These tests are generally more powerful when the assumptions are met.

Common parametric tests include:

t-tests (one-sample, independent samples, paired samples)

ANOVA (one-way, two-way, repeated measures)

Z-tests (one-sample, two-sample)

F-tests (one-way, two-way)

**Non-parametric tests**

Non-parametric tests don't assume a specific distribution of the data. They are useful when dealing with ordinal data or when the assumptions of parametric tests are violated. Examples include:

Mann-Whitney U test

Wilcoxon signed-rank test

Kruskal-Wallis test

Selecting the appropriate test

When choosing a hypothesis test, researchers consider a few broad categories:

**Data Distribution:** Determine if your data is normally distributed, as many tests assume normality.

Number of Groups: Identify how many groups you're comparing (e.g., one group, two groups, or more).

**Group Independence:** Decide if your groups are independent (different subjects) or dependent (same subjects measured multiple times).

**Data Type:**

Continuous (e.g., height, weight),

Ordinal (e.g., rankings),

Nominal (e.g., categories without order).

Based on these categories, you can select the appropriate statistical test. For instance, if your data is normally distributed and you have two independent groups with continuous data, you would use an Independent t-test. If your data is not normally distributed with two independent groups and ordinal data, a Mann-Whitney U test is recommended.

# **Modern Approaches to Hypothesis Testing**

In addition to traditional hypothesis testing methods, there are several modern approaches:

**Permutation or randomization tests**
These tests involve randomly shuffling the observed data many times to create a distribution of possible outcomes under the null hypothesis. They are particularly useful when dealing with small sample sizes or when the assumptions of parametric tests are not met.

**Bootstrapping **
Bootstrapping is a resampling technique that involves repeatedly sampling with replacement from the original dataset. It can be used to estimate the sampling distribution of a statistic and construct confidence intervals.

**Monte Carlo simulation**
Monte Carlo methods use repeated random sampling to obtain numerical results. In hypothesis testing, they can be used to estimate p-values for complex statistical models or when analytical solutions are difficult to obtain.

**Controlling for Errors**
When conducting hypothesis tests, it's best to understand and control for potential errors:

**Type I and Type II errors**
Type I Error: Rejecting the null hypothesis when it's actually true (false positive).

Type II Error: Failing to reject the null hypothesis when it's actually false (false negative).


The significance level (α) directly controls the probability of a Type I error. Decreasing α reduces the chance of Type I errors but increases the risk of Type II errors.

**To balance these errors:**

Adjust the significance level based on the consequences of each error type.
Increase sample size to improve the power of the test.
Use one-tailed tests when appropriate.

**The file drawer effect**
The file drawer effect refers to the publication bias where studies with significant results are more likely to be published than those with non-significant results. This can lead to an overestimation of effects in the literature. To mitigate this:

Consider pre-registering studies.

Publish all results, significant or not.

Conduct meta-analyses that account for publication bias.

Simulate data beforehand.

**Glossary of Key Terms and Definitions**
Null Hypothesis (H0): The default assumption that there is no effect or difference.

Alternative Hypothesis (Ha): The hypothesis that there is an effect or difference.

P-value: The probability of observing the test results under the null hypothesis.

Significance Level (α): The threshold for rejecting the null hypothesis, commonly set at 0.05.

**Test Statistic:** A standardized value used to compare the observed data with the null hypothesis.

Type I Error: Rejecting a true null hypothesis (false positive).
Type II Error: Failing to reject a false null hypothesis (false negative).
Statistical Power: The probability of correctly rejecting a false null hypothesis.

Confidence Interval: A range of values that likely contains the true population parameter.

**Effect Size:** A measure of the magnitude of the difference or relationship being tested.
